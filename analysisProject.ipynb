```python
import time
import random
import matplotlib.pyplot as plt

def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key

def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        left_half = arr[:mid]
        right_half = arr[mid:]

        merge_sort(left_half)
        merge_sort(right_half)

        i = j = k = 0

        while i < len(left_half) and j < len(right_half):
            if left_half[i] < right_half[j]:
                arr[k] = left_half[i]
                i += 1
            else:
                arr[k] = right_half[j]
                j += 1
            k += 1


def plot_time_complexity():
    sizes = [10, 50 ,100, 200, 500, 1000]
    insertion_sort_times = []
    merge_sort_times = []

    for size in sizes:


        arr_insertion=random.sample(range(size * 10), size)

        arr_sorted_insertion = arr_insertion[:]
        arr_sorted_merge = arr_insertion[:]

        arr_merge = arr_insertion.copy()

        start_time = time.time()
        insertion_sort(arr_sorted_insertion)
        insertion_sort_times.append(time.time() - start_time)

        start_time = time.time()
        merge_sort(arr_sorted_merge)
        merge_sort_times.append(time.time() - start_time)

        print(f"Un-sorted Array: {arr_insertion}")
        print(f"size: {size}, insertion_sort_time: {insertion_sort_times[-1]}, merge_sort_time: {merge_sort_times[-1]}")

        print(f"Sorted insertion arr: {arr_sorted_insertion}")
        print(f"Sorted mergeSort arr:{arr_sorted_merge}")

    plt.plot(sizes, insertion_sort_times, label='Insertion Sort')
    plt.plot(sizes, merge_sort_times, label='Merge Sort')

    for i in range(len(sizes)):
        if merge_sort_times[i] < insertion_sort_times[i]:
            plt.scatter(sizes[i], merge_sort_times[i], color='red', marker='o')
            plt.annotate(f'{sizes[i]}', (sizes[i], merge_sort_times[i]))
            break

    plt.xlabel('Array Size')
    plt.ylabel('Time (s)')
    plt.title('Time Complexity of Merge Sort and Insertion Sort')
    plt.legend()
    plt.show()

plot_time_complexity()
```
